<p>همبستگی یک اندازه بیانگر عددی از رابطه آماری بین دو متغیر است. برای ما، این متغیرها معمولاً دو ستون از داده‌ها خواهند بود، به عنوان مثال، دمای بیرون و احتمال بارندگی.</p>
<p>یکی از روش‌ها برای محاسبه همبستگی بین دو بردار داده استفاده از مقدار 
r
 پیرسون است. این مقدار بین -۱ و ۱ متغیر است. که -۱ به معنای همبستگی کاملاً منفی، ۰ به معنای عدم همبستگی، و ۱ به معنای همبستگی کاملاً مثبت است.</p>
<p>در تصویر زیر، شما می‌توانید یک نمایش تصویری از همبستگی را ببینید. </p>

<img src="statics/img/02_05_01_correlation.png" alt="Correlation" />

<p>یک نمونه از همبستگی مثبت ممکن است ارتفاع و وزن باشد زیرا به طور کلی افراد قد بلندتر وزن بیشتری دارند. برای همبستگی منفی، دمای بیرون و هزینه‌های گرمایشی زیرا هر چه بیرون گرم‌تر باشد شما کمتر از بخاری استفاده می‌کنید. برای عدم همبستگی، تعداد کفش‌هایی که دارید و ضریب هوشی شما (که کاملاً دو مورد غیر مرتبط هستند).</p>
<p>شما می‌توانید فرمول ریاضی عدد پیرسون را در <a href="https://en.wikipedia.org/wiki/Pearson_correlation_coefficient">صفحه ویکی‌پدیا</a> مربوط به آن پیدا کنید.</p>
<p>در اینجا چگونگی انجام آن با scipy آورده شده است:</p>

</div>

```python
from scipy import stats
import numpy as np

array_1 = np.array([1,2,3,4,5,6])  # ساخت آرایه Numpy
array_2 = array_1  # تولید آرایه‌ای با همان مقادیر

print(stats.pearsonr(array_1, array_2))  # محاسبه همبستگی بوسیله عدد پیرسون
```

<h2>تولید نمونه‌ها از توزیع‌ها</h2>
<p>Scipy از بسیاری از توزیع‌ها پشتیبانی می‌کند، اینجا ما به توزیع نرمال نگاه خواهیم کرد.</p>
<h3>توزیع نرمال</h3>
<p>توزیع نرمال یک توزیع است که دور میانگین متقارن است و برای آن مقادیر نزدیک‌تر به میانگین رایج‌تر هستند. </p>
<p>شما می‌توانید در تصویر ببینید که احتمالات وجود دارند که در ۱، ۲، و ۳ انحراف معیار از میانگین باشند.</p>
<p>بسیاری از پدیده‌های طبیعی به این ترتیب نرمال توزیع شده‌اند. به عنوان مثال، قد.</p>
<img src="statisc/img/02_05_02_normal_dist.png" alt="Normal Dist" />
<p>فرض کنید نیاز دارید که داده‌ها را از یک توزیع نرمال با میانگین (صفر)۰ و انحراف معیار ۱۰ تولید کنید؛ در اینجا چگونگی انجام آن آورده شده است:</p>


</div>

```python
from scipy import stats

x = stats.norm.rvs(loc=0, scale=10, size=10)  # تولید ده عدد تصادفی با میانگین صفر و انحراف از معیار ده
print(x)
```

<div dir="rtl" style="text-align: right;">

<p>کد بالا از پارامتر 
loc
 به عنوان میانگین، 
 scale
  به عنوان انحراف از معیار و 
size
 به عنوان تعداد نمونه‌های خواسته شده استفاده می‌کند. اگر شما به اندازه کافی نقطه را از این توزیع نمونه‌برداری کنید و نتایج را ترسیم کنید، شما یک توزیع نرمال متمرکز دور ۰ با انحراف معیار ۱۰ را مشاهده خواهید کرد.</p>

</div>

<h2>تابع چگالی احتمال</h2>

<h3>تابع چگالی احتمال</h3>
<p>یک عملیات رایج دیگر روی توزیع‌ها محاسبه تابع چگالی احتمال است. این تابع به شما احتمال نسبی اینکه یک مقدار خاص را نمونه‌برداری کنید را می‌دهد. بیایید به چند نمونه نگاه کنیم. </p>

</div>

```python
from scipy import stats

p1 = stats.norm.pdf(x=-100, loc=0, scale=10)  # احتمال دریافت عدد صد در توزیع احتمالی با میانگین صفر و انحراف از معیار ده
p2 = stats.norm.pdf(x=0, loc=0, scale=10)     # احتمال دریافت عدد صفر در توزیع احتمالی با میانگین صفر و انحراف از معیار ده

print(p1)
print(p2)
```

<div dir="rtl" style="text-align: right;">

<h3>تابع توزیع تجمعی</h3>
<p>یک محاسبه رایج دیگر، تابع توزیع تجمعی است، احتمال نمونه‌برداری یک مقدار کمتر یا مساوی 
x.</p>

</div>


```python
from scipy import stats

p1 = stats.norm.cdf(x=0, loc=0, scale=10)  # احتمال دریافت عددی کوچکتر مساوی صفر در این تابع توزیع احتمال

print(p1)
```

<div dir="rtl" style="text-align: right;">

<p>ما می‌توانیم ببینیم که تابع توزیع تجمعی با x=0 برابر با 0.5 است، زیرا 0 میانگین است و با یک توزیع نرمال، نیمی از داده‌ها کمتر یا مساوی میانگین هستند.</p>
<h3>محاسبه آمار توصیفی</h3>
<p>در نهایت، اگر یک آرایه از مقادیر در 
Scipy
 دارید، می‌توانید از تابع 
 describe()
  برای محاسبه چندین آمار توصیفی برای آرایه خود استفاده کنید. به برنامه زیر توجه کنید:</p>




```python
from scipy import stats

# محاسبه آمار توصیفی از ۵۰۰ داده تصادفی با میانگین صفر و انحراف از معیار ۱
print(stats.describe(stats.norm.rvs(loc=0, scale=1, size=500)))  
```
