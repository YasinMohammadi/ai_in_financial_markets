<h1>مقدمه‌ای بر مصورسازی داده</h1>

<h2>اهمیت مصورسازی داده</h2>
<p>تاکنون، ما به فهم داده‌ها از طریق آمار توصیفی و جداول پرداخته‌ایم. ابزار مفید دیگری نیز وجود دارد که مصورسازی است.</p>
<p>مصورسازی‌های داده می‌توانند مزایای زیر را فراهم کنند:</p>
<ul>
<li>درک بهتری از داده‌ها</li>
<li>داستان جذاب‌تری هنگام توضیح داده‌ها</li>
<li>رسانه‌ای آسان‌تر برای درک</li>
</ul>

<p>مصورسازی داده مهارت اصلی برای هر تحلیلگر یا دانشمند داده است. توانایی استفاده از مصورسازی‌های عالی برای کمک به روایت داستان داده‌ها اغلب به طور قابل توجهی به درک دیگران کمک می‌کند. این درک افزوده می‌تواند تفاوتی در پیشبرد یک پروژه ایجاد کند.</p>

<h2>ابزارهای مصورسازی داده در پایتون</h2>
<p>در پایتون، دو ابزار محبوب برای مصورسازی داده 
Matplotlib
 و 
 Seaborn
  هستند. این‌ها ابزارهایی هستند که ما روی آنها تمرکز خواهیم کرد، اما ابزارهای دیگری نیز وجود دارد از جمله 
  Bokeh، ggpy و D3.</p>
<p>Matplotlib
 به نوعی کتابخانه پایه ترسیم در پایتون است. آن را به عنوان یک کتابخانه سطح پایین در نظر بگیرید که به شما اجازه می‌دهد انواع کارها را انجام دهید، اما این انعطاف‌پذیری گاهی اوقات می‌تواند کار با آن را دشوار کند. 
 Matplotlib
  گاهی اوقات می‌تواند از نظر سبک کمی قدیمی به نظر برسد.</p>
<p>Seaborn
 برای کمک به حل برخی از این مشکلات ایجاد شد. این کتابخانه بر روی 
 Matplotlib
  ساخته شده است و به گفته خودش، "یک رابط سطح بالا برای رسم گرافیک‌های آماری جذاب فراهم می‌کند." بیشتر مواقع، هر زمان که ممکن است، به سمت استفاده از 
  Seaborn
   می‌روم. زمانی که به کنترل سطح پایین‌تر نیاز دارم، 
   Matplotlib
    را به کار می‌برم. از آنجایی که 
    Seaborn
     بر روی 
     Matplotlib
      بنا شده است، ترکیب این دو بسیار آسان است.</p>

<h2>مجموعه داده‌های ما</h2>

<p>شما متوجه خواهید شد که ما مجموعه داده 
Breast Cancer
 را از 
 sklean
  بارگذاری می‌کنیم تا به عنوان داده‌ها برای مصورسازی استفاده کنیم.</p>

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd

# بارگذاری داده سرطان سینه، توجه داشته باشید این داده به صورت پیش‌فرض در این کتابخانه وجود دارد
cancer_data = load_breast_cancer()
# وارد کردن داده به دیتافریم
cancer_df = pd.DataFrame(cancer_data.data, columns=cancer_data.feature_names)

# نمایش پنج سطر از داده
print(cancer_df.head())
```
