<h2>بُرداری‌سازی زمانی</h2>
<p>این درس بُرداری‌سازی زمانی را با مطالعه موردی جالبی به نام "مجموعه ماندلبروت" توضیح می‌دهد.</p>
<h3>توصیف مسئله</h3>
<p>بُرداری‌سازی زمانی جایی است که عناصر از محاسبات یکسانی استفاده می‌کنند اما نیاز به تعداد تکرارهای مختلف دارند.</p>
<p>مجموعه ماندلبروت، مجموعه‌ای از اعداد مختلط <em>c</em> است که برای آن‌ها تابع <em>f<sub>c</sub>(z) = z<sup>2</sup> + c</em> هنگام تکرار از <em>z=0</em>، منحرف نمی‌شود. یعنی دنباله‌های <em>f<sub>c</sub>(0), f<sub>c</sub>(f<sub>c</sub>(0))</em> و غیره در مقدار مطلق محدود باقی می‌ماند. محاسبه این مجموعه بسیار آسان است، اما می‌تواند زمان زیادی ببرد زیرا باید اطمینان حاصل شود که یک عدد منحرف نمی‌شود. این معمولاً با تکرار محاسبه تا حداکثر تعداد تکرارها انجام می‌شود، پس از آن، اگر عدد هنوز در محدوده‌ای باشد، به عنوان غیرمنحرف در نظر گرفته می‌شود.</p>
<p>البته، هرچه تکرارها بیشتر باشد، دقت بیشتری خواهید داشت.</p>
