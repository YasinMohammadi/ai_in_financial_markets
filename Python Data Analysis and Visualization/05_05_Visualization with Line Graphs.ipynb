<h2>نمودارهای خطی در پایتون</h2>
<p>تابع <code>lineplot()</code> از Seaborn وظیفه ترسیم نمودار خطی را بر عهده دارد. اولین پارامتر لیست یا آرایه مقادیر 
x و دومین پارامتر آرایه مقادیر 
y است.</p>
<p>همین! نمودارهای خطی برای پیاده‌سازی ساده هستند و می‌توانند بسیار مفید باشند.</p>
<p>بیایید به یک مثال با استفاده از مجموعه داده‌های پرواز از Seaborn نگاه کنیم. این مجموعه داده‌ها سه ستون دارد. سال، ماه و مسافران. ستون مسافران تعداد مسافران پرواز برای آن سال و ماه را نشان می‌دهد.</p>





<p>کد بالا مجموعه داده‌ها را بارگذاری می‌کند، داده‌ها را فقط به ماه ژانویه محدود می‌کند و سپس تعداد مسافران برای هر سال را به صورت نمودار خطی نشان می‌دهد.</p>
<p>اینگونه است که تغییرات داده‌ها در طول زمان با نمودار خطی نقشه‌برداری می‌شوند. بعد، ما به نمودارهای حرارتی نگاه خواهیم کرد.</p>
</div>


```python
import seaborn as sns           
import pandas as pd
import matplotlib.pyplot as plt
 
flights_long=sns.load_dataset("flights")   # واردکردن داده
 
# فیلتر کردن ماه ژانویه
flights_long=flights_long[flights_long.month == 'Jan']

# پیاده‌سازی نمودار خطی
plot=sns.lineplot(x='year', y='passengers', data=flights_long)
```
