<h1>مصورسازی با نمودارهای حرارتی</h1>


<h2>معرفی نمودارهای حرارتی</h2>
<p>نمودارهای حرارتی نمودارهایی هستند که داده‌ها در آن به صورت ماتریس نمایش داده می‌شوند، به صورتی که که هر مقدار در ماتریس یک دسته است.</p>
<p>به عنوان مثال، نمودار حرارتی زیر را در نظر بگیرید.</p>
<p>این نمودار حرارتی میانگین دمای ماهانه در پارک مرکزی، نیویورک را بر اساس سال (ردیف‌ها) و ماه (ستون‌ها) نمایش می‌دهد. رنگ‌آمیزی درک نمودار حرارتی آن را آسان می‌کند تا ببینیم که گرم‌ترین ماه‌ها به طور مداوم از ژوئن تا سپتامبر هستند.</p>

<h2>نمودارهای حرارتی در پایتون</h2>
<p>شما می‌توانید نمودارهای حرارتی را در پایتون با استفاده از تابع <code>heatmap()</code> از Seaborn ایجاد کنید. برخی از پارامترهای اصلی عبارتند از:</p>

<ul>
<li><code>data</code> اولین پارامتر است و ماتریس داده‌های شما به عنوان یک دیتافریم پانداس است.</li>
<li><code>annot</code> اگر True باشد، مقادیر داده‌ها را در هر خانه نمایش می‌دهد.</li>
<li><code>fmt</code> به شما اجازه می‌دهد کنترل کنید که فرمت رشته چگونه باشد. مقدار 
"d"
 از یک عدد صحیح ده‌دهی استفاده می‌کند.</li>
<li><code>linewidths</code> به شما امکان می‌دهد عرض خطوطی که هر خانه را جدا می‌کنند را تنظیم کنید.</li>
<li><code>ax</code> به شما امکان می‌دهد یک Axes سفارشی Matplotlib را ارسال کنید.</li>
<li><code>cmap</code> به شما اجازه می‌دهد رنگ‌بندی را که هنگام ترسیم استفاده می‌کنید انتخاب کنید.</li>
</ul>
<p>بیایید به یک مثال با استفاده از همان داده‌های پرواز  نگاه کنیم:</p>




</div>


```python
import seaborn as sns
import matplotlib.pyplot as plt

# خواندن داده
flights_long = sns.load_dataset("flights")
# چرخش داده به صورتی که داده طولانی به داده عرضی تبدیل شود
flights_wide = flights_long.pivot(index="month", columns="year", values="passengers")
# تولید یک شکل برای کشیدن نمودار بر روی آن
f, ax = plt.subplots(figsize=(12, 6))
# تولید نمودار حرارتی
sns.heatmap(flights_wide, annot=True, fmt="d", linewidths=.5, ax=ax, cmap='Blues')
```

<div dir="rtl" style="text-align: right;">


<p>در بالا، ما تعداد مسافران هواپیما را بر اساس ماه و سال نمایش داده‌ایم. در کد ما، ما از Matplotlib برای تنظیم اندازه تصویر بزرگتر، (12,6) استفاده کردیم. این مقداری است که ما "ax" نامیدیم و آن را با پارامتر ax به Seaborn ارسال می‌کنیم. ما همچنین از پارامتر <code>annot=True</code> برای افزودن مقادیر واقعی به خانه‌های مان استفاده کردیم.</p>
<p>گزینه دیگری که می‌خواهم به آن اشاره کنم پارامتر <code>cmap</code> است. این به ما اجازه داد تا رنگ استفاده شده برای خانه‌ها را با ارسال "Blues" تنظیم کنیم.</p>
<p> ما می‌توانیم ببینیم که در طول سال‌ها تعداد مسافران افزایش یافته است و ماه‌های جولای و آگوست به طور مداوم محبوب‌ترین ماه‌ها هستند.</p>
