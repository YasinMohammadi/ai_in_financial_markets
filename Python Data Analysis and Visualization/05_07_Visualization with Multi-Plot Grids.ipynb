<h2>معرفی شبکه‌های چند نموداری</h2>
<p>Seaborn همچنین قابلیت‌های شبکه چند نموداری دارد.</p>
<p>من فکر نمی‌کنم بتوانم آن را بهتر از مستندات 
Seaborn 
توضیح دهم:</p>
<p>هنگام بررسی داده‌های با ابعاد متوسط، یک رویکرد مفید این است که چندین نمونه از همان نمودار را در زیرمجموعه‌های مختلف داده‌های خود رسم کنید. این تکنیک گاهی اوقات به عنوان "شبکه" یا "ساختار" نامیده می‌شود و به ایده "چندگانه‌های کوچک" مرتبط است. این به بیننده امکان می‌دهد به سرعت مقدار زیادی از اطلاعات درباره داده‌های پیچیده را استخراج کند. Matplotlib پشتیبانی خوبی برای ساخت اشکال با چندین محور دارد؛ seaborn بر این اساس ساخته شده است تا ساختار نمودار را مستقیماً به ساختار داده‌های شما مرتبط کند.</p>

<h2>شبکه‌های چند نموداری در پایتون</h2>
<h3>FacetGrid و map</h3>
<p>من فکر می‌کنم ساده‌ترین راه برای درک، استفاده از یک مثال در پایتون از مجموعه داده‌های پرواز است.</p>
<p>اول، ما به ترکیب <code>FacetGrid()</code> و <code>map()</code> در Seaborn نگاه خواهیم کرد.</p>
<p>پارامترهایی که برای <code>FacetGrid</code> استفاده خواهیم کرد عبارتند از:</p>
<ul>
<li><code>data</code> که اولین پارامتر است و دیتافریمی را مشخص می‌کند که باید استفاده شود.</li>
<li><code>row</code> که نام ستون از دیتافریم شما است که می‌خواهید به عنوان ردیف‌های شبکه خود استفاده کنید.</li>
<li><code>col</code> که نام ستون از دیتافریم شما است که می‌خواهید به عنوان ستون‌های شبکه خود استفاده کنید.</li>
<li><code>margin_titles</code> که اگر True باشد، عناوین برای متغیر ردیف در سمت راست آخرین ستون کشیده می‌شوند.</li>
</ul>
<p>پارامترهایی که برای <code>map</code> استفاده خواهیم کرد عبارتند از:</p>
<ul>
<li><code>func</code> که اولین آرگومان است و تابعی است که می‌خواهید استفاده کنید.</li>
<li><code>args</code> که دومین پارامتر است و نام ستون برای متغیری است که می‌خواهید رسم کنید.</li>
<li><code>color</code> که به شما امکان می‌دهد رنگ نمودار را مشخص کنید.</li>
</ul>

</div>


```python
import seaborn as sns
import matplotlib.pyplot as plt

# واردکردن داده
flights =sns.load_dataset("flights")  
# فیلتر کردن سال‌های بعد از ۱۹۵۶
flights = flights[flights.year >= 1956]

g = sns.FacetGrid(flights, row="year", margin_titles=True)
g.map(plt.plot, "passengers", color="steelblue")
```

<div dir="rtl" style="text-align: right;">

<p>اول، ما یک <code>FacetGrid</code> را با ارسال دیتافریم خود که در این مثال flights نام دارد، ایجاد کردیم. سپس ما متغیر ردیف (year) را که می‌خواهیم در نمودار نتیجه استفاده کنیم، ارسال می‌کنیم. شما می‌توانید در نمودار بالا ببینید که ما یک نمودار برای هر سال داریم. عناوین حاشیه به نمودارها دستور می‌دهند تا عناوین را در حاشیه اضافه کنند.</p>
<p>پس از داشتن <code>FacetGrid</code> خود، ما تابع <code>map</code> را فراخوانی می‌کنیم تا داده‌ها را روی شبکه خود نگاشت کنیم. اول، به آن می‌گوییم که چه نوع نموداری می‌خواهیم استفاده کنیم. در این مثال، ما از نمودار خطی Matplotlib استفاده می‌کنیم. سپس ما متغیری را که می‌خواهیم رسم کنیم، ارسال می‌کنیم، مسافران.</p>
<p>این یک نمودار خطی برای هر ردیف برای متغیر مورد علاقه ما ایجاد می‌کند.</p>

</div>

<h3>PairGrid</h3>
<p>یکی دیگر از توابع رایج چند نموداری در Seaborn، <code>PairGrid()</code> است. این تابع روابط جفتی بین ستون‌ها را در یک شبکه نمایش می‌دهد. شما معمولاً یک پارامتر را وارد می‌کنید که دیتافریمی شامل ستون‌هایی است که می‌خواهید نمایش دهید.</p>
<p>سپس تابع <code>map()</code> را فراخوانی می‌کنید که ترکیب‌های جفتی ستون‌های شما را بر روی نمودار انتخابی شما نگاشت می‌کند. تابعی که می‌خواهید استفاده کنید، پارامتری است که به <code>map()</code> می‌دهید.</p>
<p>در اینجا یک مثال آورده شده است:</p>





<p>نتیجه، همانطور که می‌بینید، تمام نمودارهای پراکندگی جفتی متغیرهای شماست که به شما امکان می‌دهد به سرعت برای شناسایی روابط بالقوه جستجو کنید. از آنجا که قطر شبکه نمودار پراکندگی متغیرها با خودشان است، همیشه داده‌ها را روی خط ۴۵ درجه دریافت خواهید کرد.</p>
<p>شما روش‌های زیادی برای آزمایش با ترکیب <code>FacetMap</code> و <code>map</code> دارید، پس لذت ببرید! می‌توانید اطلاعات بیشتری را از مستندات Seaborn کسب کنید.</p>
<p>حالا که با تکنیک‌ها برای تجسم هر مجموعه داده‌ای آشنا شده‌اید، درس بعدی چالش‌هایی برای شما به ارمغان می‌آورد.</p>
</div>


```python
from sklearn.datasets import load_linnerud
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# استایل کردن نمودار
sns.set(style='ticks', palette='Set2')

# بارگذاری داده
linnerud_data = load_linnerud()
linnerud_df = pd.DataFrame(linnerud_data.data, columns=linnerud_data.feature_names)

# ساخت نگاشت روی گریدهایی جفتی
g = sns.PairGrid(linnerud_df[['Chins', 'Situps', 'Jumps']])
g.map(plt.scatter)

# زدودن خطوط اضافه
sns.despine()
```

<p>شما روش‌های زیادی برای آزمایش با ترکیب <code>FacetMap</code> و <code>map</code> دارید، پس لذت ببرید! می‌توانید اطلاعات بیشتری را از  <a href="https://seaborn.pydata.org/tutorial/axis_grids.html">مستندات Seaborn</a> کسب کنید.</p>
