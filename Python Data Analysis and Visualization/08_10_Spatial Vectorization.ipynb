<h2>برداری‌سازی فضایی</h2>


<h3>برداری‌سازی فضایی چیست؟</h3>

<p>بردارسازی فضایی به وضعیتی اشاره دارد که در آن عناصر یک محاسبه مشابه را به اشتراک می‌گذارند، اما فقط با یک زیرگروه از دیگر عناصر در تعامل هستند.</p>
<p>این مورد قبلاً در مثال بازی زندگی صدق می‌کرد، اما در برخی مواقع یک سختی اضافی وجود دارد زیرا زیرگروه پویا است و باید در هر تکرار به‌روزرسانی شود. این مورد، به عنوان مثال، در سیستم‌های ذرات صدق می‌کند که ذرات عمدتاً با همسایگان محلی تعامل دارند. این مورد همچنین برای "boids" که رفتارهای گله‌ای را شبیه‌سازی می‌کنند نیز صادق است.</p>

   <h2>توضیحات مسئله</h2>
    <p>Boids یک برنامه حیات مصنوعی است که توسط Craig Reynolds در سال ۱۹۸۶ توسعه یافته و رفتار گله‌ای پرندگان را شبیه‌سازی می‌کند. نام "boid" مخفف عبارت "bird-oid object" است که به یک شیء شبیه پرنده اشاره دارد. — بخشی از مقاله ویکی‌پدیا Boids</p>
    <p>همانند اکثر شبیه‌سازی‌های حیات مصنوعی، Boids مثالی از رفتار نوظهور است؛ به این معنی که پیچیدگی Boids از تعامل عوامل فردی (در این مورد، boids) که از یک مجموعه قوانین ساده پیروی می‌کنند، ناشی می‌شود. قوانینی که در ساده‌ترین دنیای Boids اعمال می‌شوند به شرح زیر است:</p>
    <ul>
        <li>جدایی: برای اجتناب از ازدحام هم‌گله‌های محلی هدایت کنید</li>
        <li>همراستایی: به سمت میانگین جهت‌گیری هم‌گله‌های محلی هدایت کنید</li>
        <li>چسبندگی: به سمت میانگین موقعیت (مرکز جرم) هم‌گله‌های محلی هدایت کنید</li>
    </ul>
    <p>Boids با مجموعه‌ای از سه قانون محلی (جدایی، چسبندگی و همراستایی) هدایت می‌شوند که به عنوان محاسبه سرعت و شتاب عمل می‌کنند.</p>
