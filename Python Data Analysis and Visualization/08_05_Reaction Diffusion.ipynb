<h3>واکنش-پخش</h3>
<h4>توضیحات مسئله</h4>
<p>واکنش و پخش گونه‌های شیمیایی می‌تواند الگوهای مختلفی داشته باشند که یادآور الگوهایی است که اغلب در طبیعت دیده می‌شوند. معادلات گری-اسکات این نوع واکنش را مدل‌سازی می‌کنند. برای اطلاعات بیشتر در مورد این سیستم شیمیایی به مقاله «الگوهای پیچیده در یک سیستم ساده» (جان ای. پیرسون، علم، جلد 261، 1993) مراجعه کنید.</p>

<p>دو گونه شیمیایی U و V با غلظت‌های مربوطه u و v و نرخ‌های پخش Du و Dv در نظر گرفته شده‌اند.</p>

<p>V با نرخ تبدیل k به P تبدیل می‌شود. f نرخ فرایندی است که U را تغذیه و U، V و P را تخلیه می‌کند. این می‌تواند به صورت زیر نوشته شود:</p>

</div>


$$  U + 2V → 3V: \quad  \dot{u} = D_u \nabla^2 u - uv^2 + f(1-u)  $$
$$V → P: \qquad   \dot{v} = D_v \nabla^2 v + uv^2 - (f+k) v $$

<h4>راه‌حل </h4>
<p>در زیر، یک سیستم واکنش-پخش را پیاده‌سازی کرده‌ایم، نگاهی به آن بیندازید! ابتدا با «باکتری 1» اجرا می‌شود، شما می‌توانید گونه‌های مختلف را با نظر دادن یا برداشتن نظر اجرا کنید!</p>




<h4>خروجی:</h4>
<p>خروجی این کد در صورت دانلود فایل به شکل زیر خواهد بود. تصویر زیر به شما کمک می‌کند تا بازی زندگی را تجسم کنید:</p>

<h4>مطالعات بیشتر:</h4>
<p>برای کسب اطلاعات بیشتر در مورد واکنش‌ها و پخش شیمیایی، می‌توانید به مقالات و کتاب‌های مرتبط در این زمینه مراجعه کنید.</p>
</div>


```python
# -----------------------------------------------------------------------------
# From Numpy to Python
# Copyright (2017) Nicolas P. Rougier - BSD license
# More information at https://github.com/rougier/numpy-book
# -----------------------------------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# پارامترها از http://www.aliensaint.com/uo/java/rd/
# -----------------------------------------------------
n = 256
Du, Dv, F, k = 0.16, 0.08, 0.035, 0.065  # باکتری 1
# Du, Dv, F, k = 0.14, 0.06, 0.035, 0.065  # باکتری 2
# (سایر تنظیمات برای باکتری‌ها، مرجان‌ها، اثر انگشت، الگوهای مارپیچی، و غیره)

Z = np.zeros((n+2, n+2), [('U', np.double),
                          ('V', np.double)])
U, V = Z['U'], Z['V']
u, v = U[1:-1, 1:-1], V[1:-1, 1:-1]

r = 20
u[...] = 1.0
U[n//2-r:n//2+r, n//2-r:n//2+r] = 0.50
V[n//2-r:n//2+r, n//2-r:n//2+r] = 0.25
u += 0.05*np.random.uniform(-1, +1, (n, n))
v += 0.05*np.random.uniform(-1, +1, (n, n))

def update(frame):
    global U, V, u, v, im

    for i in range(10):
        Lu = (                  U[0:-2, 1:-1] +
              U[1:-1, 0:-2] - 4*U[1:-1, 1:-1] + U[1:-1, 2:] +
                                U[2:  , 1:-1])
        Lv = (                  V[0:-2, 1:-1] +
              V[1:-1, 0:-2] - 4*V[1:-1, 1:-1] + V[1:-1, 2:] +
                                V[2:  , 1:-1])
        uvv = u*v*v
        # بر اساس معادلات واکنش‌های شیمیایی ذکر شده در بالا
        u += (Du*Lu - uvv + F*(1-u))
        v += (Dv*Lv + uvv - (F+k)*v)

    im.set_data(V)
    im.set_clim(vmin=V.min(), vmax=V.max())

fig = plt.figure(figsize=(4, 4))
fig.add_axes([0.0, 0.0, 1.0, 1.0], frameon=False)
im = plt.imshow(V, interpolation='bicubic', cmap=plt.cm.viridis)
plt.xticks([]), plt.yticks([])

Writer = animation.writers['ffmpeg']
writer = Writer(fps=15, metadata=dict(artist='Me'), bitrate=1800)

anim = animation.FuncAnimation(fig, update, interval=10, frames=100)
anim.save('statics/img/08_05_diffusion.mp4', writer=writer)
plt.show()
```
