<h1>داده‌های گمشده</h1>

<h2>داده‌های گمشده چیست؟</h2>
<p>اکنون شما داده‌های خود را خوانده‌اید و هدف ما این است که متوجه شویم کدام داده‌ها گم شده‌اند و باید با آن‌ها چه کرد؟ شما چه کاری انجام می‌دهید؟</p>
<p>اول، شما باید سعی کنید درک کنید که چرا نقاط داده‌های شما گم شده‌اند. آیا آنها به طور تصادفی گم شده‌اند یا خیر؟ داده‌هایی که به طور تصادفی گم می‌شوند می‌توانند با یک مجموعه داده بزرگ حذف شوند. اگر داده‌های شما به دلیلی گم شده‌اند، شاید گم شدن به معنای صفر است؟ یا آیا نشانه خوبی از خرابی سنسور است؟ این مقادیر گمشده غیر تصادفی باید بازیابی شوند یا مورد استفاده قرار گیرند. به عنوان مثال، آنها را به صفر تنظیم کنید اگر باید صفر باشند.</p>

<p>دوم، شما باید در نظر بگیرید که چقدر داده نسبت به تعداد ردیف‌هایی که داده‌های گمشده دارند، دارید. اگر ۱ میلیون نقطه داده با ۱۰ نقطه گمشده به طور تصادفی دارید، احتمالاً خوب است که آن ردیف‌ها را حذف کنید. از طرف دیگر، اگر ۳۰۰ نقطه داده با ۱۰۰ مورد گمشده دارید، پس شاید باید در مورد حذف ۱/۳ از داده‌های خود مردد باشید. با فرض اینکه نمی‌توانید دلیل زیربنایی برای داده‌های گمشده پیدا کنید، چگونه با آنها برخورد می‌کنید؟</p>

<h2>پر کردن داده‌های گمشده</h2>
<p>چند روش برای پر کردن مقادیر گمشده وجود دارد.</p>
<h3>استفاده از یک مقدار آماری</h3>
<p>مقادیر گمشده یک ستون را با یک مقدار آماری مانند میانگین، میانه یا مد پر کنید. این مزیت سادگی و آسانی انجام را دارد اما همچنین واریانس کمتری در ستون شما ایجاد می‌کند که در غیر این صورت وجود نداشت. واریانس بالا در یک ستون معمولاً برای یک مدل یادگیری ماشین مفید است (با فرض اینکه واقعی باشد). همچنین، ممکن است زیاد منطقی نباشد. به عنوان مثال، یک مجموعه داده از قد و وزن را تصور کنید که وزن‌های گمشده دارد. سپس شما تمام وزن‌های گمشده را با میانگین پر می‌کنید صرف‌نظر از قد. ما می‌دانیم که قد و وزن همبستگی دارند، اما ما از این دانش استفاده نمی‌کنیم.</p>
<p>بیایید به یک مثال ساده از ایده اول ما برای پر کردن مقادیر گمشده با یک مقدار آماری نگاه کنیم.</p>


</div>


```python
import numpy as np
import pandas as pd
pd_series = pd.Series([5, 10, np.nan, 15, 20, np.nan, 25, 50, np.nan])
print("Average of non-missing values: {0}".format(pd_series.mean()))
pd_series = pd_series.fillna(pd_series.mean())
print(pd_series)
```

<div dir="rtl" style="text-align: right;">
<p>در بالا، ما یک سری 
Pandas
 ایجاد کردیم که فقط یک ستون از یک دیتافریم 
 Pandas
  است. در سری ما، ما ۳ مقدار گمشده با 
  np.nan
   اضافه کردیم.</p>
<p>Pandas
 به شما اجازه می‌دهد به راحتی مقادیر گمشده را با تابع 
 fillna()
  پر کنید. ما این تابع را با مقداری که می‌خواهیم استفاده کنیم، می‌گذرانیم. در این مثال، ما میانگین مقادیری که گم نشده‌اند را می‌گذرانیم.</p>

</div>

<h3>استفاده از یک مدل</h3>
<p>از یک مدل برای یادگیری مقادیری که به نظر معقول برای مقادیر گمشده می‌رسند، استفاده کنید. به طور اساسی، از داده‌هایی که گم نشده‌اند برای یادگیری اینکه متغیر گمشده چه می‌تواند باشد، استفاده کنید. در مثال قد و وزن ما، شما می‌توانید از قد برای پیش‌بینی وزن زمانی که گم نشده استفاده کنید و سپس آن مدل را بر روی قدهایی با وزن‌های گمشده برای پر کردن داده‌های گمشده خود اعمال کنید. یک الگوریتم محبوب برای این کار 
K-Nearest Neighbors
 است. برای برخی ایده‌های بیشتر در مورد چگونگی رسیدگی به داده‌های گمشده، به این <a herf="https://www.linkedin.com/pulse/imputing-missing-data-playing-fire-jehan-gonsal/">مقاله</a> نگاه کنید.</p>

<h2>ردیابی و حذف داده‌های گمشده</h2>
<p>Pandas
 همچنین امکان حذف آسان مقادیر گمشده را با استفاده از تابع 
 dropna()
  فراهم می‌کند:</p>




</div>

```python
import numpy as np
import pandas as pd

# ساخت یک سری با داده گمشده
pd_series = pd.Series([5, 10, np.nan, 15, 20, np.nan, 25, 50, np.nan])
# حذف سطرهای دارای داده گمشده
pd_series = pd_series.dropna()
print(pd_series)
```

<div dir="rtl" style="text-align: right;">

<p>یا می‌توانید با تابع 
isnull()
 ببینید که کدام مقادیر گمشده‌اند.</p>


</div>

```python
import numpy as np
import pandas as pd

# ساخت یک سری با داده گمشده
pd_series = pd.Series([5, 10, np.nan, 15, 20, np.nan, 25, 50, np.nan])
# نشان دادن سطر داده‌های گمشده
print(pd_series.isnull())
```

<div dir="rtl" style="text-align: right;">



<p> یک سری جدید برگردانده می‌شود که در آن  مقدار 
True
 به این معنی است که مقدار واقعی گمشده است و  مقدار 
 False
  به این معنی است که یک مقدار واقعی وجود دارد.</p>
