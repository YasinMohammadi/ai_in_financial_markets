<p>نمودارهای پراکندگی
scatter plots
  برای نمایش دو متغیر بسیار مناسب  هستند تا ببینیم چگونه ممکن است با یکدیگر همبستگی داشته باشند و چه رابطه‌ای ممکن است بین متغیرها وجود داشته باشد.</p>
<p>به عنوان مثال، به نمودار پراکندگی زیر نگاه کنید.</p>
<p>در این نمودار، ما یک رابطه مثبت بین شعاع میانگین و میانگین نقاط مقعر می‌بینیم. دیدن این رابطه در نمودار پراکندگی با جستجوی نقاطی که نزدیک به خط ۴۵ درجه به سمت بالا و راست هستند، آسان است.</p>


<h2>نمودارهای پراکندگی در پایتون</h2>
<p>Seaborn کار ساخت نمودارهای پراکندگی را با استفاده از تابع <code>lmplot()</code> آسان می‌کند. ورودی‌های کلیدی عبارتند از:</p>
<ul>
<li><code>x</code> که نام ستون برای متغیر محور x است</li>
<li><code>y</code> که نام ستون برای متغیر محور y است</li>
<li><code>data</code> که دیتافریم پانداس شما است</li>
</ul>
<p>بیایید به این مثال نگاه کنیم:</p>

</div>

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Set the palette and style to be more minimal
sns.set(style='ticks', palette='Set2')

# Load data as explained in introductory lesson
cancer_data = load_breast_cancer()
cancer_df = pd.DataFrame(cancer_data.data, columns=cancer_data.feature_names)

# Create the scatter plot
sns.lmplot(x="mean radius", y="mean concave points", data=cancer_df)
# Remove excess chart lines and ticks for a nicer looking plot
sns.despine()
```

<div dir="rtl" style="text-align: right;">

<p>همانگونه که در کد مشاهده می‌کنید ما ما می‌خواهیم داده‌های 
x
 را به نسبت 
 y
 در نمودار نمایش دهیم.</p>
  
<p>به طور پیش‌فرض، یک خط رگرسیون در نمودار نمایش داده می‌شود تا شما بتوانید به راحتی روابط خطی بالقوه را ببینید. متغیرهای ما به نظر مثبت همبستگی دارند. سایه روشن‌تر دور خط، فاصله اطمینان ۹۵٪ برای خط رگرسیون ما است و با استفاده از بوت‌استرپ‌های داده‌های ما محاسبه می‌شود. شما می‌توانید تمام گزینه‌ها برای این نمودار را <a href="https://seaborn.pydata.org/generated/seaborn.lmplot.html">اینجا</a> ببینید.</p>

<p>همچنین، ما از یکی از سبک‌ها و پالت‌های 
seaborn
 استفاده کردیم. این به ما اجازه می‌دهد تا در سطح بالایی کنترل کنیم که نمودارهای ما چگونه به نظر می‌رسند. شما می‌توانید بیشتر در مورد کنترل زیبایی‌شناسی <a href="https://seaborn.pydata.org/tutorial/aesthetics.html#aesthetics-tutorial">اینجا</a> یاد بگیرید. ما یک سبک و پالت را انتخاب کردیم که ظاهری بیشتر مینیمالیستی ایجاد می‌کند. ما همچنین پس از ایجاد نمودار، <code>despine()</code> را فراخوانی کردیم تا خطوط و تیک‌های اضافی نمودار را حذف کنیم که به نظر من ظاهری تمیزتر ایجاد می‌کند.</p>
