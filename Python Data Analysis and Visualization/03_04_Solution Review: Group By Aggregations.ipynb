<h2>تجمیع‌های گروه‌بندی</h2>

</div>


```python
import pandas as pd

# فراخوانی داده
def read_csv():
    # تعریف نام ستون‌ها
    names = ["mpg", "cylinders", "displacement", "horsepower", "weight", "acceleration", "model_year", "origin", "car_name"]
    #خواندن داده از روی فایل
    df = pd.read_csv("data/02_12_auto-mpg.data", header=None, names=names, delim_whitespace=True)
    return df

# توصیف داده
def group_aggregation(df, group_var, agg_var):

    # گروه‌بندی داده و محاسبه میانگین
    grouped_df = df.groupby([group_var])[agg_var].mean()
    return grouped_df

# فراخوانی تابع
print(group_aggregation(read_csv(),"cylinders","mpg"))
```

<div dir="rtl" style="text-align: right;">


<p>طبق بیان مسئله، ما باید داده‌های مجموعه داده‌های مصرف سوخت خودرو را بر اساس یک ستون گروه‌بندی کنیم. سپس باید میانگین داده‌های گروه‌بندی شده را بر اساس ستون دیگری محاسبه کنیم. قبل از انجام آن، ما باید ابتدا داده‌ها را بخوانیم. نیازی به توضیح نحوه خواندن داده‌ها نیست، زیرا ما قبلاً به طور مفصل آن را مطالعه کردیم. داده‌ها از خط ۴ تا خط ۹ خوانده می‌شوند.</p>
<p> به پیاده‌سازی  هدر تابع 
group_aggregation(df, group_var, agg_var)
 نگاه کنید. این سه ورودی را به عنوان آرگومان می‌گیرد:</p>
<ul>
<li>df: یک دیتافریم شامل مجموعه داده‌ها به صورت ماتریس.</li>
<li>group_var: یک متغیر که داده‌ها را گروه‌بندی می‌کند</li>
<li>agg_var: یک متغیر برای تجمیع یا توصیف داده‌های گروه‌بندی شده با آمار</li>
</ul>
<p>خط ۱۵ مهم‌ترین خط است. ما از تابع داخلی 
groupby()
 روی df استفاده می‌کنیم که یک آرگومان می‌گیرد: 
 group_var
 . این داده‌های گروه‌بندی شده را بر اساس ستون 
 group_var
  برمی‌گرداند. سپس، ما میانگین داده‌های گروه‌بندی شده را با ستون 
  agg_var
   محاسبه می‌کنیم. سپس در خط ۱۶، نتیجه را برمی‌گردانیم.</p>

<p>در خط ۱۹ ما تابع 
group_aggregation(read_csv(), "cylinders", "mpg")
 را فراخوانی می‌کنیم. ابتدا کنترل به 
 read_csv()
  در خط ۴ منتقل می‌شود و ما یک دیتافریم دریافت می‌کنیم.</p>
<p>آرگومان بعدی 
cylinders
 است. بر اساس مجموعه داده‌ها، ما پنج مقدار مختلف برای سیلندرها داریم: ۳،۴،۵،۶، و ۸. بنابراین دیتافریم 
 df
  به پنج گروه تقسیم می‌شود، هر گروه توسط تعداد سیلندرهای مختلف نمایندگی می‌شود. به عبارت ساده، تمام اتومبیل‌های دارای ۳ سیلندر یک گروه را تشکیل می‌دهند. به همین ترتیب، تمام اتومبیل‌های دارای ۴ سیلندر یک گروه دیگر را تشکیل می‌دهند، و غیره.</p>
<p>آخرین آرگومان 
mpg
 است. بر اساس مجموعه داده‌ها، 
 mpg
  دارای مقادیر پیوسته است. بر اساس پیاده‌سازی، میانگین mpg برای تمام اتومبیل‌ها (از همان گروه) برگردانده و چاپ می‌شود.</p>
