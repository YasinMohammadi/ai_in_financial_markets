<h1>مقدمه</h1>
<h2>آنالیتیکس(Analytics) چیست؟</h2>
<p>آنالیتیکس، در اصل، فرآیندی است که طی آن داده‌های خام گرفته شده و بینش‌های قابل اقدام استخراج می‌شوند.</p>
<p>به عنوان مثال، یک مجموعه داده خام می‌تواند قیمت‌های تاریخی بازار سهام و اطلاعات مالی تمام شرکت‌های موجود در S&P 500 باشد. به خودی خود، این داده‌های خام چندان مفید نیستند. با این حال، با مهارت‌های تحلیلی، می‌توانید شروع به استخراج درک از داده‌های خود کنید. شاید بخواهید بازده متوسط بازار را بر اساس سال کشف کنید یا هر سهمی که عملکرد نامتعارفی دارد را شناسایی کنید. آنالیتیکس فرآیندی است که از طریق آن به درک داده‌های خود می‌رسید. علاوه بر این، این درک به شما امکان می‌دهد تا تصمیمات مبتنی بر داده اتخاذ کنید.</p>
<h2>فرایند آنالیتیکس</h2>
<p>دوست دارم فرایند آنالیتیکس را به چهار مرحله متمایز تقسیم کنم:</p>
<ul>
<li>پردازش داده‌(Processing the data)</li>
<li>توصیف داده‌(Describing the data)</li>
<li>پاک‌سازی داده‌( Cleaning the data)</li>
<li>تصویرسازی داده‌(Visualizing the data)</li>
</ul>
<p>معمولاً، هنگام شروع یک پروژه آنالیتیکس جدید، شما با مقدار زیادی داده و یک سوال که می‌خواهید به آن پاسخ دهید مواجه می‌شوید. ابتدا، نیاز دارید که این داده‌ها را در یک سیستم برای پردازش بخوانید. پس از بارگذاری، معمولاً شروع به محاسبه آمارها و تعدادهای مختلف برای توصیف داده‌ها می‌کنید. این فرایند معمولاً مشکلات بالقوه‌ای را آشکار می‌کند و نیاز به پاک‌سازی داده‌ها دارد. در آن نقطه، احتمالاً برمی‌گردید و تلاش بیشتری برای توصیف و درک داده‌های خود انجام می‌دهید. برای تقویت این فرایند، سپس شروع به تصویرسازی داده‌ها می‌کنید.</p>
<h2>اهمیت آنالیتیکس</h2>
<p>هدف از این فرایند کامل، شروع به کندن لایه‌های داده‌های شما برای تلاش و پاسخ به سوال یا سوالات مورد علاقه شما است. این قدرت آنالیتیکس است. این به شما امکان می‌دهد با بی‌دانشی و مقدار زیادی داده شروع کنید و به درک مبتنی بر داده برسید.</p>

<h1>چرا پایتون برای آنالیتیکس؟</h1>
<p>پایتون به سرعت به زبان انتخابی در زمینه آنالیتیکس داده، یادگیری ماشین، و علم داده تبدیل شده است.</p>
<p>سایر زبان‌ها نیز در این حوزه ارزشمند هستند، مانند R و جولیا. با این حال، به نظر من، پایتون رهبر واضح این حوزه است.</p>
<p>در واقع، در می ۲۰۱۹، R از فهرست ۲۰ زبان محبوب تیوب ایندکس خارج شد، پس از سه سال حضور در رتبه‌های برتر، در حالی که پایتون در رتبه چهارم قرار داشت و بیشترین رشد را در بین زبان‌های برتر ۲۰ تجربه کرد.</p>
<p>پایتون به دلیل قابلیت کاربرد در زمینه‌ها و منظورهای مختلف، سهولت استفاده، و اکوسیستم فوق‌العاده‌ای از کتابخانه‌ها که از آنالیتیکس داده و یادگیری ماشین پشتیبانی می‌کنند، بسیار محبوب شده است.</p>
<p>ما در این دوره از پایتون به صورت انحصاری استفاده خواهیم کرد و در مورد بسیاری از کتابخانه‌های عالی آن، مانند Numpy، Scipy، Pandas، و Seaborn بحث خواهیم کرد.</p>
