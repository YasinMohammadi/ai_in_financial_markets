<h2>بردارسازی یکنواخت</h2>


<h3>بردارسازی یکنواخت چیست؟</h3>
<p>بردارسازی یکنواخت ساده‌ترین شکل بردارسازی است که در آن تمام عناصر در هر مرحله زمانی همان محاسبات را به اشتراک می‌گذارند، بدون پردازش خاصی برای هیچ عنصری.</p>

<p>یک مورد نمونه، بازی زندگی است که توسط جان کانوی اختراع شده و یکی از اولین نمونه‌های اتوماتای سلولی است. این اتومات‌های سلولی را می‌توان به راحتی به عنوان آرایه‌ای از سلول‌ها در نظر گرفت که با مفهوم همسایه‌ها به هم متصل شده‌اند و بردارسازی آن‌ها مستقیم است.</p>
