<h3>انواع توزیع‌ها</h3>
<h4>هیستوگرام</h4>
<p>هیستوگرام، تخمینی از توزیع احتمال یک متغیر پیوسته است. برای ساختن هیستوگرام، اولین گام تقسیم دامنه مقادیر به دسته‌های متوالی و غیرمتداخل است و سپس شمارش تعداد مقادیری که در هر دسته قرار می‌گیرند. دسته‌ها معمولاً به صورت فواصل متوالی، غیرمتداخل از یک متغیر مشخص می‌شوند.</p>
<p>برای ایجاد هیستوگرام با تابع <code>displot()</code> در Seaborn نیاز به یک آرگومان دارید که متغیر پیوسته‌ای است که می‌خواهید هیستوگرام آن را بسازید.</p>
<p>هیستوگرام‌ها، اما یک پارامتر بسیار مهم دارند - تعداد دسته‌ها. شما تعداد دسته‌ها را با پارامتر دسته مشخص می‌کنید. اگر مشخص نشده باشد، Seaborn سعی می‌کند تعداد پیش‌فرضی از دسته‌ها را پیدا کند. مهم است که به یاد داشته باشید، هرچه تعداد دسته‌ها بیشتر باشد، نوسان نمودار شما بیشتر خواهد بود؛ هرچه تعداد دسته‌ها کمتر باشد، ناآگاهی بیشتری خواهد داشت. هنگام انتخاب تعداد دسته‌ها دقت کنید زیرا ممکن است نحوه دیدگاه شما نسبت به داده‌هایتان را تغییر دهد.</p>
<p>بیایید نگاهی بیندازیم:</p>

</div>


```python
from sklearn.datasets import load_wine
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# # استایل کردن نمودار
sns.set(style='ticks', palette='Set2')

# باگذاری داده
wine_data = load_wine()
wine_df = pd.DataFrame(wine_data.data, columns=wine_data.feature_names)

# ایجاد نمودار توزیع
sns.displot(wine_df['alcohol'], kde=False)
# حذف برخی خطوط و حاشیه‌ها
sns.despine()
```

<h3>نمودار جعبه‌ای</h3>
<p>Seaborn  از نمودار جعبه‌ای نیز پشتیبانی می‌کند. نمودار جعبه‌ای نمای گرافیکی از آمار خلاصه از یک توزیع است. >شما می‌توانید با استفاده از تابع <code>boxplot</code> در Seaborn یک نمودار جعبه‌ای ایجاد کنید. تنها آرگومانی که معمولاً باید ارسال کنید اولین آرگومان است که متغیر پیوسته‌ای است که می‌خواهید با نمودار جعبه‌ای نمایش دهید. در اینجا یک مثال آورده شده است:</p>




</div>


```python
from sklearn.datasets import load_wine
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# # استایل کردن نمودار
sns.set(style='ticks', palette='Set2')

# باگذاری داده
wine_data = load_wine()
wine_df = pd.DataFrame(wine_data.data, columns=wine_data.feature_names)

# ایجاد نمودار جعبه‌ای
sns.boxplot(wine_df['alcohol'], orient="h")
# حذف برخی خطوط و حاشیه‌ها
sns.despine()
```

<div dir="rtl" style="text-align: right;">







<p>خط میانی میانه است، سمت راست جعبه صدک ۷۵ام و سمت چپ صدک ۲۵ام است. ویسکلرها خطوطی هستند که از جعبه بیرون زده‌اند. معمولاً این‌ها ۱.۵ برابر محدوده بین‌چارکی (IQR) گسترش می‌یابند که IQR فاصله بین صدک‌های ۷۵ و ۲۵ است. هر مقدار خارج از ویسکلرها معمولاً به صورت نقطه‌هایی نشان داده می‌شوند و همانطور که می‌دانیم، ممکن است به عنوان ناهنجاری در نظر گرفته شوند.</p>





<h3>نمودار ویولن</h3>
<p>Seaborn یک نمودار شبیه به نمودار جعبه‌ای به نام نمودار ویولن دارد. با این نمودار شما قسمت جعبه‌ای نمودار جعبه‌ای را حذف کرده و آن را با منحنی kde که از تابع <code>displot()</code> دیدیم، جایگزین کرده‌اید. این می‌تواند بسیار مفید باشد زیرا به شما امکان می‌دهد تا قله‌های توزیع خود را ببینید. به عنوان مثال، در زیر می‌بینیم که 
flavanoids
 دو قله‌ای است. در یک نمودار جعبه‌ای، ما این تمایز را نمی‌دیدیم.</p>



</div>


```python
from sklearn.datasets import load_wine
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# # استایل کردن نمودار
sns.set(style='ticks', palette='Set2')

# باگذاری داده
wine_data = load_wine()
wine_df = pd.DataFrame(wine_data.data, columns=wine_data.feature_names)

# ایجاد نمودار ویولون
sns.violinplot(wine_df['flavanoids'], orient="h")
# حذف برخی خطوط و حاشیه‌ها
sns.despine()
```

<h3>نمودار مشترک</h3>
<p>در نهایت، نمودارهای مشترک می‌توانند هیستوگرام دو متغیر را در یک نمودار نشان دهند و همچنین نمودار پراکندگی را در میانه نشان دهند. در بالا سمت راست، شما همچنین ضریب همبستگی پیرسون بین متغیرها را به همراه مقدار p دریافت می‌کنید.</p>
<p>این کار با تابع <code>jointplot()</code> از Seaborn انجام می‌شود.</p>
<p>اولین پارامتر متغیری است که می‌خواهید روی محور x قرار دهید. دومین پارامتر متغیری است که می‌خواهید روی محور y قرار دهید.</p>
<p>همچنین می‌توانید با پارامتر kind نحوه نمایش نقاط پراکندگی را تغییر دهید. من ترجیح می‌دهم مقدار 
"hex"
 را ارسال کنم.</p>
<p>اینگونه است که داده‌های شما از طریق توزیع‌ها تجسم می‌شوند. بعد، ما به چند روش مختلف برای تجسم داده‌ها با نمودارهای خطی نگاه خواهیم کرد.</p>
</div>

```python
from sklearn.datasets import load_wine
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# # استایل کردن نمودار
sns.set(style='ticks', palette='Set2')

# باگذاری داده
wine_data = load_wine()
wine_df = pd.DataFrame(wine_data.data, columns=wine_data.feature_names)

# ایجاد نمودار مشرک
sns.jointplot(x='color_intensity', y='flavanoids', data=wine_df)
# حذف برخی خطوط و حاشیه‌ها
sns.despine()
```
