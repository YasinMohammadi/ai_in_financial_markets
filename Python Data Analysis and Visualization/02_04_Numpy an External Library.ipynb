<h1>کتابخانه Numpy</h1>
<h2>کار با آرایه‌ها </h2>
<h3>آرایه‌های تک بعدی</h3>
<p> می‌توان 
Numpy
 را به عنوان یک کتابخانه سریع و کارآمد برای ذخیره‌سازی و دستکاری داده‌ها 
- معمولاً از طریق ساختارهای داده آرایه (که شبیه به لیست‌های پایتون هستند) -
 در نظر بگیرید. شما می‌توانید تمام مستندات 
 Numpy
  را<a href="https://numpy.org/doc/stable/user/index.html">اینجا</a>  پیدا کنید.
 
   آرایه‌های 
   Numpy
    از بسیاری جهات با لیست‌های استاندارد پایتون هستند؛ به طور قابل توجه، آنها سریع‌تر هستند، فضای کمتری اشغال می‌کنند، و قابلیت‌های بیشتری دارند.  باید بدانید که این آرایه‌ها دارای اندازه و نوع داده ثابتی هستند که شما در هنگام ایجاد تعریف می‌کنید. شما نمی‌توانید بی‌نهایت مقدار جدید را مانند یک لیست به آن‌ها اضافه کنید. بگذارید به مثالی از آرایه‌های 
    Numpy 
    توجه کنیم:</p>

</div>




```python
import numpy as np

# ایجاد آرایه جدید
np_array = np.array([5, 10, 15, 20, 25, 30])
print("--0--")

# گرفتن مقادیر یکتا در آرایه
print(np.unique(np_array))
print("--1--")

# انجراف معیار آرایه
print(np.std(np_array))
print("--2--")

# بیشینه مقدار آرایه
print(np_array.max())
print("--3--")

# هر مقدار را به توان دو می‌رسانیم
print(np_array ** 2)
print("--4--")

# جمع المان‌های آرایه
print(np_array + np_array)
print("--5--")

# مجموع تمام المان‌ها پس از اینکه به توان دو رسیدند
print(np.sum(np_array ** 2))
print("--6--")

# بازگرداندن شکل آرایه: (سطر, ستون)
print(np_array.shape)
```

<h3>آرایه‌های دو بعدی</h3>
<p>پس از آرایه‌های یک بعدی باید استفاده از آرایه‌های دوبعدی که هم ردیف و هم ستون دارند، را بیاموزیم. می‌توانید ماتریس‌ها را با آرایه‌های دو بعدی نمایش دهید. نمونه‌ای از این نمایش را در کد زیر ببینید: </p>

</div>



```python
import numpy as np

# ساخت آرایه دوبعدی
print("--0--")
np_2d_array = np.array([[1,2,3], 
                        [4,5,6]])
print(np_2d_array)

# محاسبه ترانهاده ماتریس(جای سطر و ستون را عوض می‌کند.)
print("--1--")
np_2d_array_T = np_2d_array.T
print(np_2d_array_T)

# بازگرداندن شکل آرایه: (سطر, ستون)
print("--3--")
print(np_2d_array.shape)

# دست‌یابی به المان ماتریس بوسیله ایندکس
# ایندکس اول نشانگر سطر
# ایندکس دوم نشانگر سطون
# شروع شمارش ایندکس از صفر است
print("--4--")
print(np_2d_array[1,1])
print(np_2d_array[0,2])
```

<h2>قابلیت‌های مهم</h2>
<h3>محاسبه ضرب نقطه‌ای</h3>
<p>یک تابع مفید 
Numpy
 توانایی محاسبه ضرب نقطه‌ای دو بردار است. اگر به یک مرور کلی در مورد اینکه ضرب نقطه‌ای چیست نیاز دارید، می‌توانید <a href="https://en.wikipedia.org/wiki/Dot_product">صفحه ویکی‌پدیا</a> را بررسی کنید. در اینجا نحوه انجام با کد آورده شده است:</p>
</div>




```python
import numpy as np
np_array = np.array([5, 10, 15, 20, 25, 30])
dot_product = np.dot(np_array, np_array)
print(dot_product)
```

<h3>تولید مقادیر تصادفی</h3>
<p>کتابخانه
Numpy
 توابع عالی‌ای برای کمک به نمونه‌برداری تصادفی از داده‌ها دارد.</p>
<p>ابتدا، تابع 
rand()
 را بررسی می‌کنیم که یک عدد یا اعداد تصادفی بین ۰ و ۱ برمی‌گرداند. تعداد مقادیر تصادفی برگردانده شده بستگی به شکل داده شده به تابع دارد. اگر شکلی داده نشود، فقط یک عدد برگردانده می‌شود.</p>

</div>




```python
import numpy as np

# تولید یک عدد تصادفی در بازه: range: [0,1)
print("--0--")
print(np.random.rand())

# تولید یک ماتریس با آرایه‌های تصادفی در یازه range [0,1) با شکل (3,2)
print("--1--")
print(np.random.rand(3,2))
```

<div dir="rtl" style="text-align: right;">

<p>شما همچنین می‌توانید از 
randint()
 برای بازگرداندن اعداد صحیح تصادفی از کمینه مقدار (شامل) تا بیشینه مقدار (غیر شامل) استفاده کنید. شما مقادیر کمینه
 (low)
  و بیشینه
  (high)
   را به عنوان پارامترها به تابع همراه با اندازه
   (size)
    خروجی ارسال می‌کنید.</p>

</div>




```python
import numpy as np

# Low=5, High=15, Size=2. تولید دو عدد بین پنچ تا پانزده(خود پانزده تولید نمی‌شود)
print("--0--")
print(np.random.randint(5, 15, 2))

# Low=5, High=15, Size=(3,2). تولید یک ماتریس سه در دو با اعداد تصادفی بین پنج تا پانزده
print("--1--")
print(np.random.randint(5, 15, (3,2)))
```

<h3>نمونه‌برداری از داده‌ها</h3>
<p>هنگامی که داده دارید، ممکن است که بخواهید از آن‌ها نمونه‌برداری کنید.</p>
<p>به عنوان یک مرور، نمونه‌برداری روشی برای گرفتن یک گروه کوچک‌تر از یک اجتماع بزرگ‌تر است. به عنوان مثال، شما ممکن است یک نمونه تصادفی از جمعیت ایالات متحده ایجاد کنید با اینکه تصادفاً در ده خانه در شهر محل سکونت خود را بزنید. این یک نمونه از اندازه
(size)
 ۱۰ خواهد بود.</p>
<p>تابع 
choice()
 به شما امکان می‌دهد که با پاس دادن یک آرایه مشخص کنید چند مقدار باید نمونه‌برداری شود، و تصمیم بگیرید که نمونه‌برداری تکراری یا غیرتکراری انجام شود. نمونه‌برداری عیرتکراری به این معناست که یک مقدار نمی‌تواند بیش از یک بار نمونه‌برداری شود.</p>

</div>

```python
import numpy as np
array = np.array([1,2,3,4,5])

# انتخاب ده داده با تکرار
print("--0--")
print(np.random.choice(array, 10, replace=True))

# انتخاب سه داده بدون تکرار 
print("--1--")
print(np.random.choice(array, 3, replace=False))
```

<h3>مخلوط کردن تصادفی مقادیر</h3>
<p>گاهی اوقات، ممکن است بخواهید که آرایه خود را به صورت تصادفی مخلوط کنید. کتابخانه
Numpy
 تابع 
shuffle()
 را ارائه می‌دهد که دقیقاً همین کار را انجام می‌دهد.</p>

</div>

```python
import numpy as np

x = [1,2,3,4,5]  # تولید یک لیست با پنج المان
np.random.shuffle(x)  # مخلوط کردن تصادفی لیست

print(x)
```

<div dir="rtl" style="text-align: right;">

<p>همچنین، هنگام استفاده از تصادفی‌سازی، اگر می‌خواهید انتخاب‌های تصادفی هر بار یکسان باشند، نیاز دارید که یک وردی
seed
 تنظیم کنید. این با استفاده از تابع:
numpy.random.seed()
 امکان‌پذیر است. ما باید یک عدد صحیح را برای 
 seed
  کردن تصادفی‌سازی به تابع بدهیم، بنابراین نتایج قابل تکرار خواهند بود تا زمانی که همان 
  seed
   استفاده شود.</p>

</div>

```python
import numpy as np
np.random.seed(42)
```
