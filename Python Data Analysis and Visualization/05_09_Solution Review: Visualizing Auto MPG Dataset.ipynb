<p>در خط ۱۹، ما تابع <code>scatter_plot(read_csv())</code> را فراخوانی می‌کنیم. ابتدا کنترل به <code>read_csv()</code> در خط ۵ منتقل شده و ما یک دیتافریم دریافت می‌کنیم. سپس کنترل به خط ۱۳ منتقل شده و نمودار نمایش داده می‌شود.</p>
<p>نمودار به وضوح نشان‌دهنده همبستگی منفی بین این دو مقدار داده است، به این معنی که شتاب عکس مجموع جابجایی است.</p>

```python
import pandas as pd
import seaborn as sns

def read_csv():
    # تعیین نام ستون‌ها
    names = ["mpg", "cylinders", "displacement", "horsepower", "weight", "acceleration", "model_year", "origin", "car_name"]
    # خواندن فایل 
    df = pd.read_csv("statics/data/02_12_auto-mpg.data", header=None, names=names, delim_whitespace=True)
    return df

# ساخت نمودار پراکندگی
def scatter_plot(df):
    sns.lmplot(x="displacement", y="acceleration", data = df)
    # حذف خطوط اضافه
    sns.despine()

# فراخوانی تابع نمودار پراکندگی
scatter_plot(read_csv())
```

```python
import pandas as pd
import seaborn as sns

def read_csv():
    # تعیین نام ستون‌ها
    names = ["mpg", "cylinders", "displacement", "horsepower", "weight", "acceleration", "model_year", "origin", "car_name"]
    # خواندن فایل 
    df = pd.read_csv("statics/data/02_12_auto-mpg.data", header=None, names=names, delim_whitespace=True)
    return df

# ساخت نمودار میله‌ای
def bar_plot(df):
    df = df[df.car_name.str.contains('ford')]
    df = df[df["model_year"].isin([75])]

    # مقداردهی برای ساخت نمودار
    sns.barplot(x=df['car_name'], y=df['cylinders'])
    # حذف خطوط اضافه
    sns.despine()

# فراخوانی تابع
bar_plot(read_csv())
```

```python
import pandas as pd
import seaborn as sns

def read_csv():
    # تعیین نام ستون‌ها
    names = ["mpg", "cylinders", "displacement", "horsepower", "weight", "acceleration", "model_year", "origin", "car_name"]
    # خواندن فایل 
    df = pd.read_csv("statics/data/02_12_auto-mpg.data", header=None, names=names, delim_whitespace=True)
    return df

# ساخت نمودار خطی
def bar_plot(df):

    # فیلتر کردن فورد
    df=df[df.car_name.str.contains('ford')]

    # چاپ نمودار
    sns.lineplot(x=df.model_year, y=df.weight)


# فراخوانی تابع
bar_plot(read_csv())
```
